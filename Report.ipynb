```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#由于想做美股分析，因此用pandas_reader来读取股票数据,并调用雅虎API
import pandas_datareader
import datetime
import pandas_datareader.data as web
```

```python
start = datetime.datetime(2018, 1, 1)
end = datetime.datetime(2018, 11, 1)
```

```python
#软件股2018投资市场分析
salesforce = web.DataReader("CRM", 'yahoo', start, end)
atlassian = web.DataReader("TEAM", 'yahoo', start, end)
```

```python
#开盘价价格走势非常具有相似性
salesforce['Open'].plot(label='Salesforce',figsize=(16,8),title='Open Price')
atlassian['Open'].plot(label='Atlassian')
plt.legend()
```

```python
#放量交易日期目测没有相关性
salesforce['Volume'].plot(label='Salesforce',figsize=(16,8),title='Volume Traded')
atlassian['Volume'].plot(label='Atlassian')

plt.legend()
```

```python
#赛富时价格走势图
salesforce['Open']
salesforce['MA5'] = salesforce['Open'].rolling(5).mean()
salesforce['MA30'] = salesforce['Open'].rolling(30).mean()
salesforce['MA50'] = salesforce['Open'].rolling(50).mean()
salesforce[['Open','MA5','MA30','MA50']].plot(label='salesforce',figsize=(15,8));
```

```python
atlassian['Open']
atlassian['MA5'] = atlassian['Open'].rolling(5).mean()
atlassian['MA30'] = atlassian['Open'].rolling(30).mean()
atlassian['MA50'] = atlassian['Open'].rolling(50).mean()
atlassian[['Open','MA5','MA30','MA50']].plot(label='atlassian',figsize=(15,8));
```

```python
from pandas.plotting import scatter_matrix
```

```python
software_comp = pd.concat([salesforce['Open'],atlassian['Open']],axis=1)
software_comp.columns = ['Salesforce Open','Atlassian Open']
scatter_matrix(software_comp,figsize=(8,8),alpha=0.2,hist_kwds={'bins':50});
```

从散点图可判断Salesforce和Atlassian具有正的线性相关性，软件股齐头并进

#从每日价格变化来判断股价的稳定度
#如果等于0说明基本没有变化，这样的股票不具有太大的投资价值，但变化太大会导致风险过高，谨慎投资。

```python
salesforce['returns'] = salesforce['Close'].pct_change(1)
atlassian['returns'] = atlassian['Close'].pct_change(1)
```

```python
salesforce['returns'].hist(bins=50,label='Salesforce',figsize=(10,8),alpha=0.5)
atlassian['returns'].hist(bins=50,label='Atlassian',alpha=0.5)

plt.legend()
```

从价格变化来看，Atlassian的变化幅度大于Salesforce，稳定性略逊，但两股在中心的频率分布状况相似，再次证明软件股的涨跌是齐头并进的。
Atlassian的价格变动区间相比Salesforce广一些

```python
salesforce['returns'].plot(kind='kde',label='Salesforce',figsize=(12,6))
atlassian['returns'].plot(kind='kde',label='Atlassian')

plt.legend()
```

从箱图看Atlassian的回报率比Salesforce要更高一些，但是区间也更大，风险更高。
散点图再次证明了两股之间价格走势具有正相关性。

```python
box_df = pd.concat([salesforce['returns'],atlassian['returns']],axis=1)
box_df.columns = ['Salesforce Returns',' Atlassian Returns']
box_df.plot(kind='box',figsize=(8,8),colormap='jet')
#散点图
scatter_matrix(box_df,figsize=(8,8),alpha=0.2,hist_kwds={'bins':50});
```

```python
#散点图表明正相关性，两股的回报类似
box_df.plot(kind='scatter',x='Salesforce Returns',y=' Atlassian Returns',alpha=0.4,figsize=(10,8))
```

```python
salesforce['Cumulative Return'] = (1 + salesforce['returns']).cumprod()
atlassian['Cumulative Return'] = (1 + atlassian['returns']).cumprod()
```

```python
salesforce['Cumulative Return'].plot(label='Salesforce',figsize=(16,8),title='Cumulative Return')
atlassian['Cumulative Return'].plot(label='Atlassian')

plt.legend()
```
